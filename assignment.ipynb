FINDING MISSING VALUES IN COLUMNS

```python
df.isnull().sum()
```

Found that initial_price , discount and final_price are not mapped accordingly

```python
df[['initial_price','discount','final_price']]
```

Replaced missing discount values with 0 and filled other missing fields with "Unknown".

```python
for col in df.columns:
    if df[col].dtype in ['float64', 'int64']:
        df[col] = df[col].fillna(0)
    else:
        df[col] = df[col].fillna("Unknown")
```

Confirmed that the dataset contains no missing (null) values.

```python
df.isnull().sum()
```

Calculate final_price according to initial_price and discount

```python
df['final_price'] = ((df['initial_price']-(df['initial_price']*df['discount']/100)))
```
